# Vector Quantization on speech signal (signal with memory)

### Take a speech signal and read it into Python.

```python
import matplotlib.pyplot as plt
import scipy.io.wavfile as wav
import numpy as np

rate, snd = wav.read('sndfile.wav')
#Take an excerpt of 1000 samples, 
#starting at sample 2001 and plot it:
spex = snd[2000 + np.arange(1, 1000)]
plt.plot(spex)
```

```python
plt.plot(spex[2::2], spex[1::2], '+')
```

We can see: Since the odd and the even samples are similar to each other, we get a distribution of vector points near the diagonal of the space! Hence we also only need to sample this space only near the diagonal, or more generally speaking, we should sample more densely near this diagonal.
